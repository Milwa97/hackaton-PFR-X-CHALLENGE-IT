## UCZESTNICY

```python
uczestnicy['AGE'] = uczestnicy['AGE'].str.replace(',', '.').astype(float)
uczestnicy['AGE'] = (uczestnicy['AGE']- uczestnicy['AGE'].mean()) / uczestnicy['AGE'].std()
```

```python
n=30
most_frequent_nationality = uczestnicy['NATIONALITY'].value_counts()[:n].index.tolist()
print(most_frequent_nationality)
uczestnicy['NATIONALITY'] = uczestnicy['NATIONALITY'] * uczestnicy['NATIONALITY'].isin(most_frequent_nationality)
uczestnicy['NATIONALITY'].value_counts()
```

## Two different scales: POWIAT, GMINA

```python
df_POWIAT = df_inner.copy()
df_GMINA = df_inner.copy()

df_POWIAT.drop('GMINA', axis = 1, inplace = True)
df_GMINA.drop('POWIAT', axis = 1, inplace = True)
```

## ENCODE POWIAT

```python
df_POWIAT['SEX'] = df_POWIAT['SEX'].astype(object)
df_POWIAT['NATIONALITY'] = df_POWIAT['NATIONALITY'].astype(object)
df_POWIAT['PPK_BANK'] = df_POWIAT['PPK_BANK'].astype(object)
df_POWIAT['VOIVODESHIP'] = df_POWIAT['VOIVODESHIP'].astype(object)

df_POWIAT['PPK_STAGE'] = df_POWIAT['PPK_STAGE'].astype(object)
df_POWIAT['COMPANY_SIZE'] = df_POWIAT['COMPANY_SIZE'].astype(object)
df_POWIAT['COMPANY_TYPE'] = df_POWIAT['COMPANY_TYPE'].astype(object)
df_POWIAT['POWIAT'] = df_POWIAT['POWIAT'].astype(object)

categorical_bolean = (df_POWIAT.dtypes == "object").values
data_numeric = df_POWIAT.loc[:, ~categorical_bolean]
data_categorical = df_POWIAT.loc[:, categorical_bolean]
data_dummies = pd.get_dummies(data_categorical)

data_P = pd.concat([data_numeric, data_dummies], axis=1)

data_P['TARGET'] = data_P['IS_SUSPENDED']
data_P.drop('IS_SUSPENDED', axis = 1, inplace = True)
data_P.drop('MEMBER_ID', axis = 1, inplace = True)
data_P.drop('EMPL_ID', axis = 1, inplace = True)


data_P
```

## Create training / validation /testing dataset

```python
def train_validate_test_split(df, train_percent=.6, validate_percent=.2, seed=None):
    np.random.seed(seed)
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.iloc[perm[:train_end]]
    validate = df.iloc[perm[train_end:validate_end]]
    test = df.iloc[perm[validate_end:]]
    return train, validate, test
```

```python
train, validate, test = train_validate_test_split(data_P)

train.to_csv('train_set_POWIAT.csv')
validate.to_csv('val_set_POWIAT.csv')
test.to_csv('test_set_POWIAT.csv')
```
