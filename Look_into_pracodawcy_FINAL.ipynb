## UCZESTNICY

```python
uczestnicy = pd.read_csv('PPK_Uczestnicy.csv', sep=';')
uczestnicy
```

### normalize age

```python
uczestnicy['AGE'] = uczestnicy['AGE'].str.replace(',', '.').astype(float)
uczestnicy['AGE'] = (uczestnicy['AGE']- uczestnicy['AGE'].mean()) / uczestnicy['AGE'].std()
```

### Decrease number of different nationality.  

```python
print("NaN value = ", uczestnicy['NATIONALITY'].isna().sum())
uczestnicy['NATIONALITY'].value_counts()
```

```python
n=30
most_frequent_nationality = uczestnicy['NATIONALITY'].value_counts()[:n].index.tolist()
print(most_frequent_nationality)
uczestnicy['NATIONALITY'] = uczestnicy['NATIONALITY'] * uczestnicy['NATIONALITY'].isin(most_frequent_nationality)
#uczestnicy['NATIONALITY'].value_counts()
```

### Check on dates:

```python
uczestnicy.drop('WORK_START', axis = 1, inplace = True)
uczestnicy.drop('WORK_STOP', axis = 1, inplace = True)
uczestnicy.drop('CREATED_AT', axis = 1, inplace = True)
uczestnicy.drop('NUMERICAL_VALUE', axis = 1, inplace = True)
uczestnicy.drop('UOZ_START_DATE', axis = 1, inplace = True)
uczestnicy.drop('UOP_SIGN_DATE', axis = 1, inplace = True)

uczestnicy = uczestnicy.assign(RESIGNED=uczestnicy['RESIGN_DATE'].fillna(0).astype(bool))
uczestnicy['RESIGNED'] = uczestnicy['RESIGNED'].astype(int)
uczestnicy.drop('SIGN_DATE', axis = 1, inplace = True)
uczestnicy.drop('RESIGN_DATE', axis = 1, inplace = True)
```

## Pracodawcy

```python
pracodawcy = pracodawcy.assign(GMINA=pracodawcy['REGION_CODE'].str[2:])
pracodawcy = pracodawcy.assign(POWIAT=pracodawcy['REGION_CODE'].str[:2])
pracodawcy.drop( 'REGION_CODE', axis = 1, inplace = True)
pracodawcy['POWIAT'].replace("NA", "NaN", inplace=True)
pracodawcy['GMINA'].replace("NANANANA", "NaN", inplace=True)

print("VOIVODESHIP: NaN value = ", pracodawcy['VOIVODESHIP'].isna().sum())
print("POWIAT: NaN value = ", len(pracodawcy[pracodawcy['POWIAT']=="NaN"]))
print("GMINA: NaN value = ", len(pracodawcy[pracodawcy['GMINA']=="NaN"]))
```

## ADD RATIO suspended to not suspended:

```python
df_test = uczestnicy[['MEMBER_ID', 'EMPL_ID' ]]
employee = df_test.groupby('EMPL_ID').count()
employee = pd.DataFrame(employee).astype(int)
employee['EMPL_ID'] = pd.DataFrame(employee).index
employee.reset_index(drop=True, inplace=True)
employee.rename(columns={'MEMBER_ID': 'N_OF_EMPLOYEE'}, inplace=True)
employee.sort_values(by=['N_OF_EMPLOYEE'],ascending=False)
```

```python
df_inner = pd.merge(pracodawcy, employee,  left_on='ID', right_on='EMPL_ID', how='left')
df_inner.drop('EMPL_ID', axis=1, inplace=True)
df_inner.sort_values(by=['N_OF_EMPLOYEE'],ascending=False)
```

```python
df_test = uczestnicy[['IS_SUSPENDED', 'EMPL_ID' ]]

employee_not_suspended_count = df_test[df_test['IS_SUSPENDED']==0].groupby('EMPL_ID').count()
employee_not_suspended_count = pd.DataFrame(employee_not_suspended_count).astype(int)
employee_not_suspended_count['EMPL_ID'] = pd.DataFrame(employee_not_suspended_count).index
employee_not_suspended_count.reset_index(drop=True, inplace=True)
employee_not_suspended_count.rename(columns={'IS_SUSPENDED': 'IS_NOT_SUSPENDED'}, inplace=True)
employee_not_suspended_count.sort_values(by=['IS_NOT_SUSPENDED'],ascending=False)
```

```python
employee_suspended_count = df_test[df_test['IS_SUSPENDED']==1].groupby('EMPL_ID').count()
employee_suspended_count = pd.DataFrame(employee_suspended_count).astype(int)
employee_suspended_count['EMPL_ID'] = pd.DataFrame(employee_suspended_count).index
employee_suspended_count.reset_index(drop=True, inplace=True)
employee_suspended_count.sort_values(by=['IS_SUSPENDED'],ascending=False)
```

```python
df_inner = pd.merge(df_inner, employee_suspended_count,  left_on='ID', right_on='EMPL_ID', how='left')
df_inner.drop('EMPL_ID', axis=1, inplace=True)
df_inner.sort_values(by=['IS_SUSPENDED'],ascending=False)
```

```python
pracodawcy = pd.merge(df_inner, employee_not_suspended_count,  left_on='ID', right_on='EMPL_ID', how='left')
pracodawcy.drop('EMPL_ID', axis=1, inplace=True)
pracodawcy.sort_values(by=['IS_NOT_SUSPENDED'],ascending=False)

pracodawcy = pracodawcy.assign(IS_SUSPENDED=pracodawcy['IS_SUSPENDED'].fillna(0).astype(int))
pracodawcy = pracodawcy.assign(IS_NOT_SUSPENDED=pracodawcy['IS_NOT_SUSPENDED'].fillna(0).astype(int))
pracodawcy['RATIO_IS_SUSPENDED'] = pracodawcy['IS_SUSPENDED'] / pracodawcy['N_OF_EMPLOYEE']  
pracodawcy
```

```python
tmp = np.array(pracodawcy['RATIO_IS_SUSPENDED'])
tmp = np.where(tmp > 0.7, 1, tmp)
tmp = np.where(tmp < 0.3, 0, tmp)
tmp = np.where( abs(tmp-0.5)<0.3, np.nan, tmp)

pracodawcy['RATIO_IS_SUSPENDED'] = tmp

print("NaN value = ", pracodawcy['RATIO_IS_SUSPENDED'].isna().sum())
pracodawcy['RATIO_IS_SUSPENDED'].value_counts()
```

```python
pracodawcy.drop('N_OF_EMPLOYEE', axis=1, inplace=True)
pracodawcy.drop('IS_SUSPENDED', axis=1, inplace=True)
pracodawcy.drop('IS_NOT_SUSPENDED', axis=1, inplace=True)
pracodawcy.head(20)
```

## Create training / validation /testing dataset

```python
def train_validate_test_split(df, train_percent=.6, validate_percent=.2, seed=None):
    np.random.seed(seed)
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.iloc[perm[:train_end]]
    validate = df.iloc[perm[train_end:validate_end]]
    test = df.iloc[perm[validate_end:]]
    return train, validate, test
```

```python
train, validate, test = train_validate_test_split(data_V)

train.to_csv('train_set_V.csv',index = False)
validate.to_csv('val_set_V.csv',index = False)
test.to_csv('test_set_V.csv',index = False)
```
